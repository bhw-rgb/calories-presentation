```python
# 완전한 EDA 및 머신러닝 파이프라인: 운동 데이터 분석
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import skew, boxcox
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor, ExtraTreesClassifier
from sklearn.cluster import DBSCAN
from sklearn.metrics import f1_score
import warnings
warnings.filterwarnings('ignore')

# 데이터 로드
df = pd.read_csv('train.csv')

# 기본 정보 확인
print("=" * 60)
print("데이터셋 기본 정보")
print("=" * 60)
print(f"데이터 크기: {df.shape}")
print(f"\n컬럼명:\n{df.columns.tolist()}")
print(f"\n데이터 타입:\n{df.dtypes}")
print(f"\n첫 5행:\n{df.head()}")
print(f"\n데이터 정보:\n{df.info()}")

# 새로운 피처 생성
print("\n" + "=" * 60)
print("새로운 피처 생성")
print("=" * 60)

# BMI 계산 (if height and weight columns exist)
if 'Height' in df.columns and 'Weight' in df.columns:
    df['BMI'] = df['Weight'] / ((df['Height'] / 100) ** 2)
    print("✓ BMI 피처 생성")

# 칼로리 대비 운동 시간 효율성
if 'Calories' in df.columns and 'Duration' in df.columns:
    df['Calorie_Efficiency'] = df['Calories'] / (df['Duration'] + 1)  # +1 to avoid division by zero
    print("✓ 칼로리 효율성 피처 생성")

# 심박수 범위 (BPM 관련)
if 'Heart_Rate' in df.columns:
    df['HR_Zone'] = pd.cut(df['Heart_Rate'],
                           bins=[0, 100, 140, 170, 220],
                           labels=['Rest', 'Light', 'Moderate', 'Intense'])
    print("✓ 심박수 존 피처 생성")

# 나이 그룹
if 'Age' in df.columns:
    df['Age_Group'] = pd.cut(df['Age'],
                             bins=[0, 30, 40, 50, 100],
                             labels=['Young', 'Middle', 'Senior', 'Elder'])
    print("✓ 나이 그룹 피처 생성")

print(f"\n최종 피처 수: {df.shape[1]}")
print(f"새로 생성된 피처들:\n{[col for col in df.columns if col not in ['ID', 'Age', 'Gender', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp', 'Calories', 'Weight_Status']]}")

print("\n" + "=" * 60)
print("타겟 변수와 피처 간의 관계 분석")
print("=" * 60)

# 타겟 변수 설정 (예: Weight_Status 또는 Calories)
# Weight_Status가 있다면 분류 문제, 없다면 Calories를 타겟으로 회귀 문제
target_column = 'Weight_Status' if 'Weight_Status' in df.columns else 'Calories'

print(f"\n타겟 변수: {target_column}")

# 타겟 변수 분포 확인
if df[target_column].dtype == 'object':
    # 범주형 타겟
    print(f"\n타겟 변수 분포:\n{df[target_column].value_counts()}")
    print(f"\n타겟 변수 비율:\n{df[target_column].value_counts(normalize=True)}")

    # 시각화
    plt.figure(figsize=(10, 5))
    df[target_column].value_counts().plot(kind='bar', color='steelblue')
    plt.title(f'{target_column} 분포', fontsize=14, fontweight='bold')
    plt.xlabel(target_column)
    plt.ylabel('빈도')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    # 수치형 타겟
    print(f"\n타겟 변수 기술통계:\n{df[target_column].describe()}")

    # 시각화
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # 히스토그램
    axes[0].hist(df[target_column].dropna(), bins=30, color='steelblue', edgecolor='black')
    axes[0].set_title(f'{target_column} 분포', fontsize=14, fontweight='bold')
    axes[0].set_xlabel(target_column)
    axes[0].set_ylabel('빈도')

    # Boxplot
    axes[1].boxplot(df[target_column].dropna(), vert=True)
    axes[1].set_title(f'{target_column} Boxplot', fontsize=14, fontweight='bold')
    axes[1].set_ylabel(target_column)

    plt.tight_layout()
    plt.show()

print("\n" + "=" * 60)
print("수치형 변수 기술통계량")
print("=" * 60)

# 수치형 컬럼 선택
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"\n수치형 컬럼 ({len(numeric_cols)}개): {numeric_cols}")

# 기술통계량
numeric_stats = df[numeric_cols].describe().T
numeric_stats['range'] = numeric_stats['max'] - numeric_stats['min']
numeric_stats['iqr'] = numeric_stats['75%'] - numeric_stats['25%']
numeric_stats['cv'] = (numeric_stats['std'] / numeric_stats['mean']) * 100  # 변동계수

print(f"\n상세 기술통계량:\n{numeric_stats}")

# 시각화
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

metrics = ['mean', 'std', 'range', 'cv']
colors = ['steelblue', 'coral', 'mediumseagreen', 'plum']

for idx, (metric, color) in enumerate(zip(metrics, colors)):
    numeric_stats[metric].plot(kind='barh', ax=axes[idx], color=color)
    axes[idx].set_title(f'{metric.upper()} 비교', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('값')

plt.tight_layout()
plt.show()

print("\n" + "=" * 60)
print("범주형 변수 기술통계량")
print("=" * 60)

# 범주형 컬럼 선택 (object 타입)
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"\n범주형 컬럼 ({len(categorical_cols)}개): {categorical_cols}")

if len(categorical_cols) > 0:
    for col in categorical_cols:
        print(f"\n{'=' * 40}")
        print(f"변수명: {col}")
        print(f"{'=' * 40}")
        print(f"고유값 개수: {df[col].nunique()}")
        print(f"\n빈도 분포:\n{df[col].value_counts()}")
        print(f"\n비율 분포 (%):\n{df[col].value_counts(normalize=True) * 100}")

    # 시각화: 범주형 변수 분포
    n_cols = len(categorical_cols)
    if n_cols > 0:
        fig, axes = plt.subplots(1, min(n_cols, 3), figsize=(15, 5))
        if n_cols == 1:
            axes = [axes]

        for idx, col in enumerate(categorical_cols[:3]):  # 최대 3개까지만 표시
            df[col].value_counts().plot(kind='bar', ax=axes[idx], color='steelblue')
            axes[idx].set_title(f'{col} 분포', fontsize=12, fontweight='bold')
            axes[idx].set_xlabel(col)
            axes[idx].set_ylabel('빈도')
            axes[idx].tick_params(axis='x', rotation=45)

        plt.tight_layout()
        plt.show()
else:
    print("범주형 변수가 없습니다.")

print("\n" + "=" * 60)
print("수치형 Feature 분포 시각화")
print("=" * 60)

# 실수형과 정수형 컬럼만 선택
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
n_features = len(numeric_cols)

if n_features > 0:
    # 그리드 크기 계산
    n_rows = (n_features + 2) // 3
    n_cols = min(3, n_features)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten() if n_features > 1 else [axes]

    for idx, col in enumerate(numeric_cols):
        axes[idx].hist(df[col].dropna(), bins=30, color='steelblue', edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'{col} 분포', fontsize=11, fontweight='bold')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('빈도')
        axes[idx].grid(alpha=0.3)

    # 남은 서브플롯 숨기기
    for idx in range(n_features, len(axes)):
        axes[idx].axis('off')

    plt.tight_layout()
    plt.show()
else:
    print("수치형 변수가 없습니다.")

print("\n" + "=" * 60)
print("수치형 Feature Boxplot - 이상치 확인")
print("=" * 60)

if n_features > 0:
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
    axes = axes.flatten() if n_features > 1 else [axes]

    for idx, col in enumerate(numeric_cols):
        axes[idx].boxplot(df[col].dropna(), vert=True, patch_artist=True,
                         boxprops=dict(facecolor='lightblue', color='steelblue'),
                         medianprops=dict(color='red', linewidth=2),
                         whiskerprops=dict(color='steelblue'),
                         capprops=dict(color='steelblue'))
        axes[idx].set_title(f'{col} Boxplot', fontsize=11, fontweight='bold')
        axes[idx].set_ylabel(col)
        axes[idx].grid(alpha=0.3, axis='y')

    # 남은 서브플롯 숨기기
    for idx in range(n_features, len(axes)):
        axes[idx].axis('off')

    plt.tight_layout()
    plt.show()

from statsmodels.stats.outliers_influence import variance_inflation_factor

print("\n" + "=" * 60)
print("다중공선성 분석 (VIF)")
print("=" * 60)

# 수치형 변수만 선택 (결측치 제거)
numeric_df = df[numeric_cols].dropna()

# VIF 계산
if numeric_df.shape[1] > 1:
    vif_data = pd.DataFrame()
    vif_data['Feature'] = numeric_df.columns
    vif_data['VIF'] = [variance_inflation_factor(numeric_df.values, i)
                       for i in range(numeric_df.shape[1])]
    vif_data = vif_data.sort_values('VIF', ascending=False)

    print("\nVIF 값 (높을수록 다중공선선 문제):")
    print("  - VIF < 5: 문제 없음")
    print("  - 5 ≤ VIF < 10: 주의")
    print("  - VIF ≥ 10: 다중공선선 문제")
    print(f"\n{vif_data}")

    # 시각화
    plt.figure(figsize=(10, 6))
    colors = ['red' if x >= 10 else 'orange' if x >= 5 else 'green' for x in vif_data['VIF']]
    plt.barh(vif_data['Feature'], vif_data['VIF'], color=colors)
    plt.xlabel('VIF 값', fontsize=12)
    plt.title('변수별 다중공선성 (VIF)', fontsize=14, fontweight='bold')
    plt.axvline(x=5, color='orange', linestyle='--', label='주의 (VIF=5)')
    plt.axvline(x=10, color='red', linestyle='--', label='문제 (VIF=10)')
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("VIF 계산을 위한 충분한 변수가 없습니다.")

# 상관관계 행렬
print("\n" + "=" * 60)
print("상관관계 분석")
print("=" * 60)

correlation_matrix = numeric_df.corr()
print(f"\n상관관계 행렬:\n{correlation_matrix}")

# 히트맵
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('변수 간 상관관계 히트맵', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 높은 상관관계 쌍 찾기
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            high_corr_pairs.append({
                'Feature_1': correlation_matrix.columns[i],
                'Feature_2': correlation_matrix.columns[j],
                'Correlation': correlation_matrix.iloc[i, j]
            })

if high_corr_pairs:
    high_corr_df = pd.DataFrame(high_corr_pairs).sort_values('Correlation',
                                                              ascending=False,
                                                              key=abs)
    print(f"\n높은 상관관계 쌍 (|r| > 0.7):\n{high_corr_df}")
else:
    print("\n높은 상관관계를 가진 변수 쌍이 없습니다.")

print("\n" + "=" * 60)
print("결측치 분석")
print("=" * 60)

# 결측치 개수
missing_counts = df.isnull().sum()
missing_percent = (missing_counts / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_counts,
    'Percentage': missing_percent
}).sort_values('Missing_Count', ascending=False)

print(f"\n결측치 현황:\n{missing_df[missing_df['Missing_Count'] > 0]}")

# 결측치 시각화
if missing_df['Missing_Count'].sum() > 0:
    plt.figure(figsize=(12, 6))
    missing_plot_data = missing_df[missing_df['Missing_Count'] > 0]
    plt.bar(missing_plot_data.index, missing_plot_data['Percentage'], color='coral')
    plt.xlabel('변수명', fontsize=12)
    plt.ylabel('결측 비율 (%)', fontsize=12)
    plt.title('변수별 결측치 비율', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("\n✓ 결측치가 없습니다.")

print("\n" + "=" * 60)
print("이상치 처리 (IQR 방법)")
print("=" * 60)

def detect_outliers_iqr(df, column):
    """IQR 방법으로 이상치 탐지"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# 이상치 탐지 및 제거
df_no_outliers = df.copy()
outlier_summary = []

for col in numeric_cols:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    n_outliers = len(outliers)
    outlier_percent = (n_outliers / len(df)) * 100

    outlier_summary.append({
        'Feature': col,
        'N_Outliers': n_outliers,
        'Percentage': outlier_percent,
        'Lower_Bound': lower,
        'Upper_Bound': upper
    })

    # 이상치 제거
    df_no_outliers = df_no_outliers[
        (df_no_outliers[col] >= lower) & (df_no_outliers[col] <= upper)
    ]

outlier_summary_df = pd.DataFrame(outlier_summary)
print(f"\n이상치 요약:\n{outlier_summary_df}")

print(f"\n원본 데이터 크기: {df.shape}")
print(f"이상치 제거 후 크기: {df_no_outliers.shape}")
print(f"제거된 행 수: {df.shape[0] - df_no_outliers.shape[0]}")

print("\n" + "=" * 60)
print("왜도(Skewness) 분석")
print("=" * 60)

# 왜도 계산
skewness_values = df[numeric_cols].skew().sort_values(ascending=False)
print(f"\n변수별 왜도:\n{skewness_values}")
print("\n해석:")
print("  - |Skewness| < 0.5: 거의 대칭")
print("  - 0.5 ≤ |Skewness| < 1: 약간 치우침")
print("  - |Skewness| ≥ 1: 매우 치우침")

# 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 왜도 막대그래프
colors = ['red' if abs(x) >= 1 else 'orange' if abs(x) >= 0.5 else 'green'
          for x in skewness_values]
axes[0].barh(skewness_values.index, skewness_values.values, color=colors)
axes[0].set_xlabel('왜도', fontsize=12)
axes[0].set_title('변수별 왜도', fontsize=14, fontweight='bold')
axes[0].axvline(x=0, color='black', linestyle='-', linewidth=0.8)
axes[0].axvline(x=-1, color='red', linestyle='--', alpha=0.5)
axes[0].axvline(x=1, color='red', linestyle='--', alpha=0.5)

# 가장 치우친 변수의 분포
most_skewed_col = skewness_values.abs().idxmax()
axes[1].hist(df[most_skewed_col].dropna(), bins=30, color='steelblue', edgecolor='black')
axes[1].set_title(f'{most_skewed_col} 분포\n(왜도: {skewness_values[most_skewed_col]:.2f})',
                  fontsize=12, fontweight='bold')
axes[1].set_xlabel(most_skewed_col)
axes[1].set_ylabel('빈도')

plt.tight_layout()
plt.show()

print("\n" + "=" * 60)
print("데이터 변환 (Log, Box-Cox, 제곱근)")
print("=" * 60)

# 변환할 변수 선택 (왜도가 높은 변수)
transform_cols = [col for col in numeric_cols if abs(df[col].skew()) > 1]

if len(transform_cols) > 0:
    transform_col = transform_cols[0]  # 첫 번째 변수로 예시
    print(f"\n변환 예시: {transform_col} (원본 왜도: {df[transform_col].skew():.2f})")

    # 양수로 변환 (Box-Cox는 양수 필요)
    data = df[transform_col].dropna()
    if (data <= 0).any():
        data = data - data.min() + 1

    # 변환 적용
    log_transform = np.log1p(data)
    sqrt_transform = np.sqrt(data)
    boxcox_transform, lambda_param = boxcox(data)

    # 왜도 비교
    print(f"\n변환 후 왜도:")
    print(f"  - 로그 변환: {skew(log_transform):.2f}")
    print(f"  - 제곱근 변환: {skew(sqrt_transform):.2f}")
    print(f"  - Box-Cox 변환: {skew(boxcox_transform):.2f} (λ={lambda_param:.2f})")

    # 시각화
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    axes[0, 0].hist(data, bins=30, color='steelblue', edgecolor='black')
    axes[0, 0].set_title(f'원본 (Skew: {skew(data):.2f})', fontweight='bold')
    axes[0, 0].set_xlabel(transform_col)

    axes[0, 1].hist(log_transform, bins=30, color='coral', edgecolor='black')
    axes[0, 1].set_title(f'로그 변환 (Skew: {skew(log_transform):.2f})', fontweight='bold')
    axes[0, 1].set_xlabel(f'log({transform_col})')

    axes[1, 0].hist(sqrt_transform, bins=30, color='mediumseagreen', edgecolor='black')
    axes[1, 0].set_title(f'제곱근 변환 (Skew: {skew(sqrt_transform):.2f})', fontweight='bold')
    axes[1, 0].set_xlabel(f'sqrt({transform_col})')

    axes[1, 1].hist(boxcox_transform, bins=30, color='plum', edgecolor='black')
    axes[1, 1].set_title(f'Box-Cox 변환 (Skew: {skew(boxcox_transform):.2f})', fontweight='bold')
    axes[1, 1].set_xlabel(f'boxcox({transform_col})')

    plt.tight_layout()
    plt.show()
else:
    print("왜도가 높은 변수가 없어 변환을 수행하지 않습니다.")

print("\n" + "=" * 60)
print("Z-Score 기반 이상치 검출")
print("=" * 60)

def detect_outliers_zscore(df, column, threshold=3):
    """Z-Score 방법으로 이상치 탐지"""
    z_scores = np.abs(stats.zscore(df[column].dropna()))
    outlier_indices = np.where(z_scores > threshold)[0]
    return outlier_indices, z_scores

# Z-Score 이상치 검출
zscore_outlier_summary = []

for col in numeric_cols:
    outlier_idx, z_scores = detect_outliers_zscore(df, col)
    n_outliers = len(outlier_idx)
    outlier_percent = (n_outliers / len(df)) * 100

    zscore_outlier_summary.append({
        'Feature': col,
        'N_Outliers': n_outliers,
        'Percentage': outlier_percent
    })

zscore_summary_df = pd.DataFrame(zscore_outlier_summary)
print(f"\nZ-Score 이상치 요약 (|Z| > 3):\n{zscore_summary_df}")

# 시각화: 첫 번째 변수의 Z-Score 분포
if len(numeric_cols) > 0:
    sample_col = numeric_cols[0]
    _, z_scores = detect_outliers_zscore(df, sample_col)

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(range(len(z_scores)), z_scores, alpha=0.5, s=10)
    plt.axhline(y=3, color='r', linestyle='--', label='임계값 (±3)')
    plt.axhline(y=-3, color='r', linestyle='--')
    plt.xlabel('데이터 인덱스')
    plt.ylabel('Z-Score')
    plt.title(f'{sample_col} Z-Score 분포', fontweight='bold')
    plt.legend()
    plt.grid(alpha=0.3)

    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=50, color='steelblue', edgecolor='black')
    plt.axvline(x=3, color='r', linestyle='--', label='임계값 (±3)')
    plt.axvline(x=-3, color='r', linestyle='--')
    plt.xlabel('Z-Score')
    plt.ylabel('빈도')
    plt.title(f'{sample_col} Z-Score 히스토그램', fontweight='bold')
    plt.legend()

    plt.tight_layout()
    plt.show()

print("\n" + "=" * 60)
print("DBSCAN 기반 이상치 탐지")
print("=" * 60)

# 수치형 데이터만 사용
from sklearn.preprocessing import StandardScaler

X_dbscan = df[numeric_cols].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_dbscan)

# DBSCAN 적용
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

# 이상치는 -1로 라벨링됨
n_outliers = np.sum(clusters == -1)
n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)

print(f"\nDBSCAN 결과:")
print(f"  - 클러스터 수: {n_clusters}")
print(f"  - 이상치 수: {n_outliers}")
print(f"  - 이상치 비율: {(n_outliers / len(X_dbscan)) * 100:.2f}%")

# 시각화 (처음 2개 변수 사용)
if len(numeric_cols) >= 2:
    plt.figure(figsize=(10, 6))

    # 정상 데이터
    normal_mask = clusters != -1
    plt.scatter(X_scaled[normal_mask, 0], X_scaled[normal_mask, 1],
                c=clusters[normal_mask], cmap='viridis',
                alpha=0.6, s=30, label='정상')

    # 이상치
    outlier_mask = clusters == -1
    plt.scatter(X_scaled[outlier_mask, 0], X_scaled[outlier_mask, 1],
                c='red', marker='x', s=100, label='이상치', linewidths=2)

    plt.xlabel(f'{numeric_cols[0]} (표준화)', fontsize=12)
    plt.ylabel(f'{numeric_cols[1]} (표준화)', fontsize=12)
    plt.title('DBSCAN 이상치 탐지', fontsize=14, fontweight='bold')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

print("\n" + "=" * 60)
print("Train/Test 데이터 분할")
print("=" * 60)

# 타겟 변수 설정
target_col = 'Weight_Status' if 'Weight_Status' in df.columns else 'Calories'

# 피처와 타겟 분리
X = df.drop(columns=[target_col], errors='ignore')
y = df[target_col] if target_col in df.columns else None

# 범주형 변수 임시 제거 (나중에 인코딩)
X_numeric = X.select_dtypes(include=[np.number])

# Train/Test 분할 (80:20)
if y is not None:
    X_train, X_test, y_train, y_test = train_test_split(
        X_numeric, y, test_size=0.2, random_state=42, stratify=y if y.dtype == 'object' else None
    )

    print(f"\n데이터 분할 완료:")
    print(f"  - 전체 데이터: {len(df)} 샘플")
    print(f"  - 학습 데이터: {len(X_train)} 샘플 ({len(X_train)/len(df)*100:.1f}%) ")
    print(f"  - 테스트 데이터: {len(X_test)} 샘플 ({len(X_test)/len(df)*100:.1f}%) ")
    print(f"  - 피처 수: {X_train.shape[1]}")
else:
    print("타겟 변수를 찾을 수 없습니다.")

print("\n" + "=" * 60)
print("RandomForest 기반 결측치 대체")
print("=" * 60)

def impute_with_rf(df, target_col, n_estimators=100):
    """RandomForest로 결측치 대체"""
    df_imputed = df.copy()

    if df[target_col].isnull().sum() > 0:
        # 결측치가 있는 행과 없는 행 분리
        train_data = df[df[target_col].notnull()]
        test_data = df[df[target_col].isnull()]

        # 피처 선택 (target_col 제외한 수치형 변수)
        feature_cols = [col for col in df.select_dtypes(include=[np.number]).columns
                       if col != target_col and df[col].isnull().sum() == 0]

        if len(feature_cols) > 0 and len(train_data) > 0:
            # RandomForest 학습
            rf = RandomForestRegressor(n_estimators=n_estimators, random_state=42, n_jobs=-1)
            rf.fit(train_data[feature_cols], train_data[target_col])

            # 결측치 예측
            predictions = rf.predict(test_data[feature_cols])
            df_imputed.loc[df[target_col].isnull(), target_col] = predictions

            print(f"✓ {target_col}: {len(test_data)}개 결측치 대체 완료")
        else:
            print(f"✗ {target_col}: 대체 불가 (피처 부족)")

    return df_imputed

# 결측치가 있는 컬럼에 대해 대체
df_imputed = df.copy()
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df_imputed = impute_with_rf(df_imputed, col)

print(f"\n결측치 대체 후:")
print(f"  - 결측치 총 개수: {df_imputed.isnull().sum().sum()}")

print("\n" + "=" * 60)
print("범주형 데이터 인코딩 (LabelEncoder)")
print("=" * 60)

df_encoded = df_imputed.copy()
label_encoders = {}

# Changed to include 'category' dtype
categorical_cols = df_encoded.select_dtypes(include=['object', 'category']).columns.tolist()

if len(categorical_cols) > 0:
    for col in categorical_cols:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = le

        print(f"✓ {col} 인코딩 완료: {len(le.classes_)}개 클래스")
        print(f"   매핑: {dict(zip(le.classes_, le.transform(le.classes_)))}")

    print(f"\n총 {len(categorical_cols)}개 범주형 변수 인코딩 완료")
else:
    print("인코딩할 범주형 변수가 없습니다.")

# 인코딩 후 데이터 확인
print(f"\n인코딩 후 데이터 타입:\n{df_encoded.dtypes}")

print("\n" + "=" * 60)
print("이상치 제거 및 Z-Score 표준화")
print("=" * 60)

# 타겟 변수 설정 (위에서 정의된 target_column 사용)
target_col = 'Weight_Status' if 'Weight_Status' in df_encoded.columns else 'Calories'

# 피처와 타겟 분리 (df_encoded에서 분리)
X_preproc = df_encoded.drop(columns=[target_col], errors='ignore')
y_preproc = df_encoded[target_col]

# 피처의 수치형 컬럼만 선택 (아웃라이어 제거 및 스케일링 대상)
numeric_features_cols = X_preproc.select_dtypes(include=[np.number]).columns.tolist()

# Outlier removal based on Z-score (applying to X and Y jointly by filtering indices)
# Create a mask for non-outliers in features
non_outlier_mask = pd.Series(True, index=X_preproc.index)
for col in numeric_features_cols:
    if col in X_preproc.columns:
        # Z-score calculation should handle NaN if present, but previous steps suggest no NaNs
        z_scores = np.abs(stats.zscore(X_preproc[col]))
        outlier_indices_for_col = X_preproc.index[z_scores >= 3]
        non_outlier_mask.loc[outlier_indices_for_col] = False

X_clean = X_preproc[non_outlier_mask]
y_clean = y_preproc[non_outlier_mask]

print(f"이상치 제거:")
print(f"  - 원본 크기: {len(df_encoded)}")
print(f"  - 이상치 제거 후: {len(X_clean)} (피처) / {len(y_clean)} (타겟)")
print(f"  - 제거된 행: {len(df_encoded) - len(X_clean)}")

# Z-Score 표준화 (피처 X_clean에만 적용)
scaler = StandardScaler()
X_scaled = X_clean.copy() # Make a copy to avoid SettingWithCopyWarning
X_scaled[numeric_features_cols] = scaler.fit_transform(X_clean[numeric_features_cols])

# For the summary, define df_scaled as the combined X_scaled and y_clean
df_scaled = X_scaled.copy()
df_scaled[target_col] = y_clean # Add back the original (integer) target for shape/dtype consistency

print(f"\n✓ Z-Score 표준화 완료 ({len(numeric_features_cols)}개 피처)")
print(f"\n표준화 후 통계량 (평균≈0, 표준편차≈1) (피처):")
print(X_scaled[numeric_features_cols].describe().loc[['mean', 'std']])

print("\n" + "=" * 60)
print("ExtraTreesClassifier 모델 학습 및 평가")
print("=" * 60)

# 타겟 변수 설정 (분류 문제)
# target_col은 이미 'Weight_Status'로 설정되었고, y_clean은 int64 타입입니다.
# df_scaled를 X_scaled와 y_clean으로 직접 대체합니다.

# 데이터 준비 (X_scaled와 y_clean 사용)
X_model = X_scaled
y_model = y_clean

if target_col and not y_model.empty:
    # Train/Test 분할
    X_train, X_test, y_train, y_test = train_test_split(
        X_model, y_model, test_size=0.2, random_state=42, stratify=y_model
    )

    print(f"데이터 분할:")
    print(f"  - 학습 데이터: {X_train.shape}")
    print(f"  - 테스트 데이터: {X_test.shape}")
    print(f"  - 클래스 수: {len(np.unique(y_model))}")

    # ExtraTreesClassifier 학습
    print("\n모델 학습 중...")
    etc = ExtraTreesClassifier(
        n_estimators=100,
        max_depth=20,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=42,
        n_jobs=-1
    )
    etc.fit(X_train, y_train)

    # 예측
    y_pred_train = etc.predict(X_train)
    y_pred_test = etc.predict(X_test)

    # Macro F1 Score 계산
    train_f1 = f1_score(y_train, y_pred_train, average='macro')
    test_f1 = f1_score(y_test, y_pred_test, average='macro')

    print("\n" + "=" * 60)
    print("모델 성능 (Macro F1 Score)")
    print("=" * 60)
    print(f"  - 학습 데이터 Macro F1: {train_f1:.4f}")
    print(f"  - 테스트 데이터 Macro F1: {test_f1:.4f}")
    print(f"  - 과적합 정도: {abs(train_f1 - test_f1):.4f}")

    # Feature Importance
    feature_importance = pd.DataFrame({
        'Feature': X_model.columns,
        'Importance': etc.feature_importances_
    }).sort_values('Importance', ascending=False)

    print(f"\n변수 중요도 Top 10:")
    print(feature_importance.head(10))

    # 시각화
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # F1 Score 비교
    scores = [train_f1, test_f1]
    labels = ['Train', 'Test']
    colors = ['steelblue', 'coral']
    axes[0].bar(labels, scores, color=colors, edgecolor='black', linewidth=1.5)
    axes[0].set_ylabel('Macro F1 Score', fontsize=12)
    axes[0].set_title('ExtraTreesClassifier 성능', fontsize=14, fontweight='bold')
    axes[0].set_ylim([0, 1])
    for i, v in enumerate(scores):
        axes[0].text(i, v + 0.02, f'{v:.4f}', ha='center', fontweight='bold')

    # Feature Importance
    top_features = feature_importance.head(10)
    axes[1].barh(top_features['Feature'], top_features['Importance'], color='mediumseagreen')
    axes[1].set_xlabel('중요도', fontsize=12)
    axes[1].set_title('변수 중요도 Top 10', fontsize=14, fontweight='bold')
    axes[1].invert_yaxis()

    plt.tight_layout()
    plt.show()

else:
    print("분류를 위한 타겟 변수가 없거나, 데이터가 비어있어 모델 학습을 수행하지 않습니다.")
    test_f1 = 0.0 # Ensure test_f1 is defined for the summary

print("\n" + "=" * 60)
print("전체 분석 파이프라인 요약")
print("=" * 60)

summary = f"""
✅ 완료된 분석 단계:

1. ✓ 데이터 정보 확인 및 피처 생성
2. ✓ 피처와 타겟 변수 관계 분석
3. ✓ 수치형 데이터 기술통계량
4. ✓ 범주형 데이터 기술통계량 및 분포
5. ✓ 수치형 Feature 분포 (히스토그램)
6. ✓ 수치형 Feature Boxplot (이상치 확인)
7. ✓ 다중공선성(VIF) 및 상관관계 분석
8. ✓ 결측치 분석
9. ✓ 이상치 처리 (IQR 방법)
10. ✓ 데이터 변환 및 왜도 시각화
11. ✓ 로그/Box-Cox/제곱근 변환
12. ✓ Z-Score 이상치 검출
13. ✓ DBSCAN 이상치 탐지
14. ✓ Train/Test 데이터 분할
15. ✓ RandomForest 결측치 대체
16. ✓ LabelEncoder 범주형 인코딩
17. ✓ 이상치 제거 및 Z-Score 표준화
18. ✓ ExtraTreesClassifier Macro F1 Score 계산

📈 최종 결과:
- 원본 데이터: {df.shape[0]}행 × {df.shape[1]}열
- 전처리 후: {df_scaled.shape[0]}행 × {df_scaled.shape[1]}열
- 제거된 행: {df.shape[0] - df_scaled.shape[0]}
- Macro F1 Score (Test): {test_f1:.4f}

🎯 핵심 인사이트:
- 모든 EDA 및 전처리 단계가 체계적으로 수행됨
- 다양한 이상치 탐지 방법 적용 (IQR, Z-Score, DBSCAN)
- 데이터 변환으로 왜도 개선
- ExtraTreesClassifier로 최종 분류 모델 구축 완료
"""

print(summary)
# 위 모든 코드를 하나의 Python 스크립트로 실행하면 완전한 EDA 및 머신러닝 파이프라인이 자동으로 수행됩니다. 각 단계는 시각화와 함께 상세한 분석 결과를 제공하며, 최종적으로 ExtraTreesClassifier의 Macro F1 Score를 계산하여 모델 성능을 평가합니다.
# • ✅ 18단계 완전 자동화: 데이터 로드부터 모델 평가까지
# • ✅ 다양한 이상치 탐지: IQR, Z-Score, DBSCAN
# • ✅ 데이터 변환: 로그, Box-Cox, 제곱근
# • ✅ 결측치 처리: RandomForest 기반 고급 대체
# • ✅ 다중공선성 분석: VIF 및 상관관계 매트릭스
# • ✅ 시각화: 각 단계마다 시각적 분석 제공
# • ✅ 모델 평가: Macro F1 Score 및 Feature Importance
# 이 파이프라인은 운동 데이터뿐만 아니라 다른 데이터셋에도 적용 가능하도록 일반화되어 있습니다.
# 또는
```
